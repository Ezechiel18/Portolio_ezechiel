#Analyse spatiale de l'accessibilité à l'eau  dans la commune Ogou 1 au Togo

L’objectif principal de l’étude est d’analyser la répartition et l’accessibilité aux infrastructures 
d’eau potable dans la commune d’Ogou 1 afin d’identifier les inégalités existantes et de 
proposer des priorités d’action pour améliorer l’accès.
